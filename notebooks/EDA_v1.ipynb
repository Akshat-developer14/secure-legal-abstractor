    Extracting useful columns and save it in json format with chat like interface for fine tuning llm

```python
# Variables required for fine tuning
target_pairs = {
    "Document Name": "Document Name-Answer",
    "Parties": "Parties-Answer",
    "Governing Law": "Governing Law-Answer",
    "Effective Date": "Effective Date-Answer",
    "Termination For Convenience": "Termination For Convenience-Answer"
}

jsonl_records = []

# Loop through the entire dataset row by row
for idx, row in data.iterrows():
    
    for clause_name, answer_column in target_pairs.items():
        raw_clause_text = str(row.get(clause_name, ""))
        clean_answer = str(row.get(answer_column, ""))
        
        # Skip rows where the legal answer is blank, null, or missing
        if pd.isna(row[answer_column]) or clean_answer.strip() in ["", "[]", "No", "None"]:
            continue
            
        # Create the conversational training structure for the LLM
        system_prompt = (
            f"You are a secure, local corporate legal assistant. Analyze the provided contract text "
            f"to isolate information regarding the following clause: '{clause_name}'. "
            f"Extract the exact details and output them in a structured JSON schema."
        )
        
        user_content = f"Contract Snippet:\n{raw_clause_text}\n\nTask: Extract the exact value for '{clause_name}'."
        
        # Clean up any residual bracket formatting from the raw dataset strings
        cleaned_value = clean_answer.replace("[", "").replace("]", "").replace("'", "").replace('"', "").strip()
        
        assistant_json_output = {
            "clause_identified": clause_name,
            "extracted_value": cleaned_value
        }
        
        # Format explicitly as a ChatML message loop
        message_block = {
            "messages": [
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_content},
                {"role": "assistant", "content": json.dumps(assistant_json_output)}
            ]
        }
        
        jsonl_records.append(message_block)

# Save the cleaned and extracted dataset to a json file
output_file = "training_data_of_legal_documents.jsonl"
with open(output_file, "w", encoding="utf-8") as f:
    for record in jsonl_records:
        f.write(json.dumps(record) + "\n")
        
print(f"\nSuccess! Generated {len(jsonl_records)} training conversation blocks from the full dataset.")
print(f"Saved file as: {output_file}")
```
