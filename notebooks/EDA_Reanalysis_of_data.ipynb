    Trying to remove some noisy data we don't require for the purpose

```python
data = data.drop(columns=['Renewal Term','Renewal Term-Answer','Notice Period To Terminate Renewal','Notice Period To Terminate Renewal- Answer'])
```

    Making null values as 'Not found' so model don't break during fine tuning but also do not hallucinate that there can be null values possible

```python
null_value_cols = ['Parties-Answer', 'Agreement Date-Answer', 'Governing Law-Answer', 'Effective Date-Answer', 'Expiration Date-Answer']

fill_dict = {col: 'Not found' for col in null_value_cols}

data = data.fillna(value=fill_dict)
```

```python
print(data.isnull().sum().to_string())
```

    Droping all the question columns that has no use in fine tuning

```python
cols_to_drop = ['Document Name', 'Parties', 'Agreement Date', 'Effective Date', 'Expiration Date', 'Governing Law', 'Most Favored Nation', 'Competitive Restriction Exception', 'Non-Compete', 'Exclusivity', 'No-Solicit Of Customers', 'No-Solicit Of Employees', 'Non-Disparagement', 'Termination For Convenience', 'Rofr/Rofo/Rofn', 'Change Of Control', 'Anti-Assignment', 'Revenue/Profit Sharing', 'Price Restrictions', 'Minimum Commitment', 'Volume Restriction', 'Ip Ownership Assignment', 'Joint Ip Ownership', 'License Grant', 'Non-Transferable License', 'Affiliate License-Licensor', 'Affiliate License-Licensee', 'Unlimited/All-You-Can-Eat-License', 'Irrevocable Or Perpetual License', 'Source Code Escrow', 'Post-Termination Services', 'Audit Rights', 'Uncapped Liability', 'Cap On Liability', 'Liquidated Damages', 'Warranty Duration', 'Insurance', 'Covenant Not To Sue', 'Third Party Beneficiary']

data = data.drop(columns=cols_to_drop)
```

```python
data.columns
```

    Striping answer from last so model to misunderstand

```python
data.columns = data.columns.str.strip().str.removesuffix('-Answer')

print(data.columns)
```

```python
data.shape
```

    Drop columns which is not much useful and create problem for model

```python
cols_not_needed = ['Source Code Escrow', 'Rofr/Rofo/Rofn', 'Covenant Not To Sue', 'Joint Ip Ownership', 'Most Favored Nation', 'Competitive Restriction Exception', 'Third Party Beneficiary', 'Non-Transferable License', 'Affiliate License-Licensor', 'Affiliate License-Licensee', 'Unlimited/All-You-Can-Eat-License', 'Irrevocable Or Perpetual License', 'Minimum Commitment', 'Volume Restriction']

data = data.drop(columns=cols_not_needed)
```

    Keeping most useful columns & also removing the boolean columns

```python
columns_to_keep = [
    'Document Name', 
    'Parties', 
    'Agreement Date', 
    'Effective Date', 
    'Expiration Date', 
    'Governing Law', 
    'Text'
]

data = data[columns_to_keep]

print("Final Training Columns:\n", data.columns)
```
